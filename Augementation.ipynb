```python
import os
from PIL import Image
from torchvision import transforms

# Define the directory containing the original spectrogram images
original_dir = '/content/drive/MyDrive/Deep Learning/Spectrogram/rock'

# Define the directory to save the augmented images
augmented_dir = '/content/drive/MyDrive/Deep Learning/Augumented Spectogram/rock'

# Create the augmented directory if it doesn't exist
os.makedirs(augmented_dir, exist_ok=True)

# Define the transformation for horizontal flipping
flip_transform = transforms.RandomHorizontalFlip(p=1.0)

# Iterate over the original images
for image_name in os.listdir(original_dir):
    # Load the original image
    image_path = os.path.join(original_dir, image_name)
    image = Image.open(image_path)

    # Apply horizontal flipping
    flipped_image = flip_transform(image)

    # Save the flipped image
    flipped_image_name = 'flipped_' + image_name
    flipped_image_path = os.path.join(augmented_dir, flipped_image_name)
    flipped_image.save(flipped_image_path)

    # Save the original image as well
    original_image_path = os.path.join(augmented_dir, image_name)
    image.save(original_image_path)
```
